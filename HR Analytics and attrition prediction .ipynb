```python
#Overall Attrition Count
sns.countplot(x="Attrition" , data = df, color = "Blue")
plt.title("Employee Attrition Count")
plt.show()
```

```python
#Department Wise Attrition
plt.figure(figsize=(8,5))
sns.countplot(x="Department", hue="Attrition", data=df, palette = ['Red', 'Blue'])
plt.title("Department vs Attrition")
plt.show()
```

```python
#Salary Vs Attrition
plt.figure(figsize=(7,5))
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
#Attrition by Agegroup
grouped = df.groupby("Agegroup")["Attrition"].value_counts(normalize=True).mul(100).rename("Rate").reset_index()
sns.barplot(x="Agegroup", y= "Rate", hue = "Attrition", data=grouped)
plt.title("Attrition Based on Age Group")
plt.show()
```

```python
#Attrition by WorkLife balance
sns.countplot(x="WorkLifeBalance", hue="Attrition", data=df)
plt.title("Work-Life Balance Impact on Attrition")
plt.show()
```

```python
#Attrition Vs Jobrole
plt.figure(figsize=(30,10))
sns.countplot(x = "JobRole", hue = "Attrition", data = df, palette = ["red", "blue"])
plt.title("Attrition By JobRole")
plt.show()
```
